```python
# Function to calculate GC content
def gc_content(sequence):
    return ((sequence.count('G') + sequence.count('C')) / len(sequence)) * 100

# Example DNA sequence
sequence = "ATGCATATGCGCGCGCGCGCGCGGATTTTTTTTGGGGGCCCCCCCATAAGGGGCCCCCGCTTACG"
print(f"GC Content: {gc_content(sequence):.2f}%")
```

```python
import matplotlib.pyplot as plt

# Data for GC content in different samples
gc_contents = [40, 50, 60, 55, 65]
samples = ['Sample A', 'Sample B', 'Sample C', 'Sample D', 'Sample E']

# Plot the bar chart
plt.bar(samples, gc_contents, color='skyblue')
plt.xlabel('Samples')
plt.ylabel('GC Content (%)')
plt.title('GC Content Across Samples')
plt.show()
```

```python
#Michaelis-Menten simulation In enzymology, the Michaelis-Menten equation describes the rate of enzymatic reactions depending on the concentration of the substrate.
substrate_concentration = 0.1
km = 0.5
vmax = 1.2
#The equation describes how the reaction rate increases as the substrate concentration increases but also plateaus as it approaches the maximum rate.

while substrate_concentration <= 1.0:
    rate = (vmax * substrate_concentration) / (km + substrate_concentration)
    print(f"[S]: {substrate_concentration}, Rate: {rate:.2f}")
    substrate_concentration += 0.1
```

```python
from Bio import pairwise2

# Example sequences
seq1 = "ATCGT"
seq2 = "ATGGT"

# Perform a simple pairwise alignment
alignments = pairwise2.align.globalxx(seq1, seq2)

for alignment in alignments:
    print(f"Alignment: {alignment[0]}")
    print(f"Score: {alignment[2]}")
```

```python
from Bio import Align

# Example sequences
seq1 = "ATCGT"
seq2 = "ATGGT"

# Create a PairwiseAligner object
aligner = Align.PairwiseAligner()

# Set the match and mismatch scores (optional)
aligner.match_score = 2
aligner.mismatch_score = -1

# Perform the alignment
alignments = aligner.align(seq1, seq2)

# Print the alignments and scores
for alignment in alignments:
    print(f"Alignment: {alignment}")
    print(f"Score: {alignment.score}")
```
