```python
# Biological Data Formats (FASTA, GenBank, etc.)
from Bio import SeqIO
from Bio import Entrez
# Fetch the GenBank file for NM_000546.6 (TP53)
handle = Entrez.efetch(db="nucleotide", id="NM_000546.6", rettype="gb", retmode="text")
record = SeqIO.read(handle, "genbank")
handle.close()

# Save the GenBank file
SeqIO.write(record, "tp53_sequence.gb", "genbank")

# Print basic annotation info
print("\n--- Sequence Annotation ---")
print(f"ID: {record.id}")
print(f"Name: {record.name}")
print(f"Description: {record.description}")
print(f"Length: {len(record.seq)} bp")
print(f"Number of Features: {len(record.features)}")
print(f"Organism: {record.annotations['organism']}")
print(f"Date: {record.annotations['date']}")

# Print gene features
print("\n--- Gene Features ---")
for feature in record.features:
    if feature.type == "gene":
        print(f"Gene: {feature.qualifiers.get('gene', [''])[0]}")
        print(f"Location: {feature.location}")
        print(f"Qualifiers: {feature.qualifiers}")
        print()
```

```python
# Install required packages
!pip install biopython matplotlib

from Bio import Phylo
from Bio.Phylo.TreeConstruction import DistanceTreeConstructor, DistanceMatrix
import matplotlib.pyplot as plt

# Create a sample distance matrix for a simple phylogenetic tree
names = ["A", "B", "C", "D"]
matrix = [
    [0],       # A
    [5, 0],    # B
    [9, 10, 0],# C
    [7, 8, 4, 0] # D
]

# Convert list to BioPython DistanceMatrix
dm = DistanceMatrix(names, matrix)

# Construct the tree using UPGMA (you can change to 'nj' for Neighbor-Joining)
constructor = DistanceTreeConstructor()
tree = constructor.upgma(dm)

# Draw the phylogenetic tree
Phylo.draw(tree)
plt.show()
```

```python
from Bio import PDB
from Bio.PDB import PDBList
import pandas as pd

# Download the PDB file
pdbl = PDBList()
pdbl.retrieve_pdb_file("1TIM", pdir=".", file_format="pdb")  # Downloads to current directory

# Example: Parsing the downloaded PDB file
parser = PDB.PPBuilder()
structure = PDB.PDBParser().get_structure("1TIM", "pdb1tim.ent")  # Use downloaded file name

# Assuming you want to get the structure of the first two chains (modify as needed)
chains_of_interest = ['A', 'B']  # Replace with actual chain IDs if different

data = []  # List to store the data

for model in structure:
    for chain in model:
        if chain.id in chains_of_interest:
            for residue in chain:
                for atom in residue:
                    row = {
                        'Chain': chain.id,
                        'Residue': residue.get_resname(),
                        'Residue ID': residue.id,
                        'Atom': atom.get_name(),
                        'Coordinates': atom.get_coord()
                    }
                    data.append(row)

df = pd.DataFrame(data)  # Create a pandas DataFrame
display(df)  # Display the DataFrame as a table
```

```python
from Bio import Entrez
Entrez.email = "......@gmail.com"  # Replace with your email

# Step 1: Search for the human gene TPI1
search_handle = Entrez.esearch(db="gene", term="1TIM[Gene Name] AND Homo sapiens[Organism]")
record = Entrez.read(search_handle)
search_handle.close()

# Check if any results were found
if record["IdList"]:
    human_gene_id = record["IdList"][0]
    print(f"Human 1TIM Gene ID: {human_gene_id}")

    # Step 2: Fetch the full gene record to check for orthologs
    fetch_handle = Entrez.efetch(db="gene", id=human_gene_id, rettype="xml")
    data = Entrez.read(fetch_handle)
    fetch_handle.close()

    # Step 3: Extract ortholog information (if available)
    # Check if 'Homologs' key exists before accessing it
    if "Homologs" in data[0]:
        homologs = data[0]["Homologs"]

        print("\n--- Orthologs for Human 1TIM ---")
        for organism in homologs:
            for ortholog in organism.get("Gene", []):
                name = ortholog.get("Gene-ref", {}).get("Gene-ref_locus", "N/A")
                species = organism.get("TaxName", "Unknown species")
                print(f"{species}: {name}")
    else:
        print("\n--- No ortholog information found for Human 1TIM ---")
else:
    print("No matching gene IDs found for 1TIM in Homo sapiens.")
```
